```python
import pandas as pd
```

```python
# Reading csvs
departments=pd.read_csv("data/departments.csv")
dept_emp=pd.read_csv("data/dept_emp.csv")
dept_manager=pd.read_csv("data/dept_manager.csv")
employees=pd.read_csv("data/employees.csv")
salaries=pd.read_csv("data/salaries.csv")
titles=pd.read_csv("data/titles.csv")

```

```python
# List the employee number, last name, first name, sex, and salary of each employee (2 points)
emp_with_salaries=employees.merge(salaries, on="emp_no", how="inner", suffixes=("_emp", "_salary"))
reduced_emp_with_salaries=emp_with_salaries[['emp_no','last_name','first_name', 
       'sex', 'salary']]

reduced_emp_with_salaries
```

```python
# List the first name, last name, and hire date for the employees who were hired in 1986 (2 points)
emp_with_salaries['hire_date'] = pd.to_datetime(emp_with_salaries['hire_date'])
emp_with_salaries_1986 = emp_with_salaries.loc[emp_with_salaries['hire_date'].dt.year == 1986]
emp_with_salaries_1986=emp_with_salaries_1986[['first_name', 'last_name', 'hire_date']]
emp_with_salaries_1986
```

```python
# List the manager of each department along with their department number, department name, employee number, 
# last name, and first name (2 points)
man_details=dept_manager.merge(departments, on="dept_no", how="inner", suffixes=("_managers", "_departments"))
man_details=man_details.merge(employees,on="emp_no", how="inner", suffixes=("_manag", "_employees"))
man_details=man_details[['dept_no','dept_name', 'emp_no','last_name', 'first_name']]
man_details
```

```python
# List the department number for each employee along with that employee’s employee number, 
#last name, first name, and department name (2 points)
emp_with_departments=employees.merge(dept_emp, on="emp_no", how="inner", suffixes=("_emp", "_dept_emp"))

def department(row):
    department = row["dept_no"]
    
    if department == "d001":
        return "Marketing"
    elif department == "d002":
        return "Finance"
    elif department == "d003":
        return "Human Resources"
    elif department == "d004":
        return "Production"
    elif department == "d005":
        return "Development"
    elif department == "d006":
        return "Quality Management"
    elif department == "d007":
        return "Sales"
    elif department == "d008":
        return "Research"
    elif department == "d009":
        return "Customer Service"
      
emp_with_departments["Department name"] = emp_with_departments.apply(department, axis=1)
emp_with_departments_reduced=emp_with_departments[['emp_no','last_name','first_name', 'Department name']]
emp_with_departments_reduced
```

```python
# List first name, last name, and sex of each employee whose first name is Hercules and whose last name begins with the letter B (2 points)
hercules_b = employees.loc[(employees['first_name'] == "Hercules") & (employees['last_name'].str.startswith("B"))]
hercules_b=hercules_b[['first_name', 'last_name','sex']]
hercules_b
```

```python
# List each employee in the Sales department, including their employee number, last name, and first name (2 points)
employees_sales= emp_with_departments.loc[emp_with_departments["Department name"]=="Sales"]
employees_sales=employees_sales[['Department name','emp_no','last_name','first_name']]
employees_sales
```

```python
# List each employee in the Sales and Development departments, including their employee number, last name, first name, and department name (4 points)
employees_sales_development = emp_with_departments.loc[(emp_with_departments["Department name"] == "Sales") | (emp_with_departments["Department name"] == "Development")]
employees_sales_development=employees_sales_development[['Department name','emp_no','last_name','first_name']]
employees_sales_development
```

```python
# List the frequency counts, in descending order, of all the employee last names 
# (that is, how many employees share each last name) (4 points)
last_names_employees = employees.groupby(['last_name']).count()[['emp_no']]
last_names_employees.sort_values('emp_no', ascending=False)
```
